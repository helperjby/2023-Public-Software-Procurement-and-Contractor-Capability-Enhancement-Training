```python
def plot_categorical_data(df, column):
    # 데이터 집계
    count_data = df[column].value_counts()
    total = count_data.sum()
    percentage_data = (count_data / total) * 100

    # 플롯 생성
    plt.figure(figsize=(10, 6))
    plot = sns.countplot(x=column, data=df, order=count_data.index)

    # 각 막대에 레이블 추가
    for i, bar in enumerate(plot.patches):
        height = bar.get_height()
        label = f"{count_data[i]} ({percentage_data[i]:.1f}%)"
        plot.text(bar.get_x() + bar.get_width()/2., height, label, ha='center', va='bottom')

    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=15)
    plt.show()
```

```python
columns = ['소속기관의 유형을 체크해주세요', '귀하의 발주 경험을 입력해주세요.', '귀하의 발주 횟수를 입력해주세요.']

# 각 컬럼에 대해 시각화 실행
for column in columns:
    plot_categorical_data(df, column)
```

```python
def plot_boxplots(df, columns):
    # Boxplot을 그리기 위한 데이터 형태로 변환
    melted_data = pd.melt(df, value_vars=columns)
    
    # 시각화
    plt.figure(figsize=(15, 8))
    sns.boxplot(x='variable', y='value', data=melted_data)
    plt.title('Boxplot of Survey Questions')
    plt.xlabel('Question')
    plt.ylabel('Scores')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()  # 레이블이 겹치지 않도록 레이아웃 조정
    plt.show()
```

```python
def plot_average_scores(df, group_by_column, title):
    q_columns = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5']
    average_scores = df.groupby(group_by_column)[q_columns].mean().reset_index()

    # 플롯을 위한 데이터 재구조화
    melted_data = pd.melt(average_scores, id_vars=[group_by_column], value_vars=q_columns)

    # 시각화
    plt.figure(figsize=(14, 6))
    plot = sns.barplot(x=group_by_column, y='value', hue='variable', data=melted_data)
    
    # y축 범위 설정
    plt.ylim(50, 100)

    # 각 막대에 레이블 추가
    for p in plot.patches:
        plot.annotate(f"{p.get_height():.1f}", 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='center', 
                      xytext=(0, 10), 
                      textcoords='offset points')

    plt.title(title, y=1.05)
    plt.xlabel(group_by_column)
    plt.ylabel('Average Score')
    plt.xticks(rotation=15)
    plt.legend(title='Question', loc='upper right')
    plt.show()
```

```python
plot_average_scores(df, '소속기관의 유형을 체크해주세요', '기관 유형별 평균 역량향상도 점수')

# 2. '귀하의 발주 경험을 입력해주세요.' 별 시각화
plot_average_scores(df, '귀하의 발주 경험을 입력해주세요.', '발주 경력 연차별 평균 역량향상도 점수')

# 3. '귀하의 발주 횟수를 입력해주세요.' 별 시각화
plot_average_scores(df, '귀하의 발주 횟수를 입력해주세요.', '발주 경력 횟수별 평균 역량향상도 점수')
```

```python
def plot_countplots(df, column):
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(data=df, x=column, palette='viridis', order=range(1, 8))
    
    # 각 막대 위에 값을 표시
    total = len(df[column])  # 전체 응답 수
    for p in ax.patches:
        count = p.get_height()
        percentage = f'{100 * count / total:.1f}%'
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(f'{int(count)}\n({percentage})', (x, y), ha='center', va='bottom')
    
    plt.title(f'{column}')
    plt.xlabel('Scale Value')
    plt.ylabel('Count')
    plt.ylim(0, df[column].value_counts().max() + 5)  # y축의 상한을 조정
    plt.show()

# 각 컬럼에 대해 함수를 실행
for col in columns_to_plot:
    plot_countplots(df, col)
```
